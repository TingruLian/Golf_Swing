```python
%matplotlib inline
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
dataset = pd.read_excel('golf_good.xlsx')
```

```python
dfname=pd.ExcelFile('golf_good.xlsx')
print(dfname.sheet_names)
dfname1=pd.ExcelFile('golf_bad.xlsx')
print(dfname1.sheet_names)

```

```python
dfnew=pd.read_excel('golf_good.xlsx',sheet_name='good2')
dfnew1 = np.linalg.norm(dfnew,axis=0)
print('new',dfnew/dfnew1)
```

```python
for items in dfname.sheet_names[:]:
  dfnew=pd.read_excel('golf_good.xlsx',sheet_name=items)#print(dfnew.describe())
  x=dfnew.iloc[:, 0]
  y=dfnew.iloc[:, 1]
  plt.scatter(x, y, label=items)
for items in dfname1.sheet_names[:]:
  dfnew1=pd.read_excel('golf_bad.xlsx',sheet_name=items)#print(dfnew.describe())
  x1=dfnew1.iloc[:, 0]
  y1=dfnew1.iloc[:, 1]
  plt.scatter(x1, y1, label=items)
  
plt.title('Centroid Location ')
plt.xlabel('X Coordinates')
plt.ylabel('Y Coordinates')
plt.legend();
```

```python
for items in dfname.sheet_names[:]:
  dfnew=pd.read_excel('golf_good.xlsx',sheet_name=items)#print(dfnew.describe())
  dfnew =np.linalg.norm(dfnew,axis=0)/dfnew
  x=dfnew.iloc[:, 0]
  y=dfnew.iloc[:, 1]
  plt.scatter(x, y, label=items)
for items in dfname1.sheet_names[:]:
  dfnew1=pd.read_excel('golf_bad.xlsx',sheet_name=items)#print(dfnew.describe())
  dfnew1 =np.linalg.norm(dfnew,axis=0)/dfnew
  x1=dfnew1.iloc[:, 0]
  y1=dfnew1.iloc[:, 1]
  plt.scatter(x1, y1, label=items)
  
plt.title('Normalized Centroid Location ')
plt.xlabel('X Coordinates')
plt.ylabel('Y Coordinates')
plt.legend();
```
